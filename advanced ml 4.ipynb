```python
def de_cipher_simple(text, char_to_count, result=True):
    cnt = CountVectorizer(analyzer='char')
    coc = cnt.fit_transform([text])
    char_occur = dict(zip(cnt.get_feature_names(), coc.sum(axis=0).tolist()[0])) 
    sorted_sec = [x[0] for x in sorted(list(char_occur.items()), key = lambda x: x[1])][::-1]
    de_cipher = dict(zip(sorted_sec, list(char_to_count.keys())[::-1][:len(sorted_sec)]))
    res = ''.join([de_cipher[x] for x in text])
    print()
    print('РАСШИФРОВКА')
    if result:
        return res
    else:
        return de_cipher
```

```python
def visual_similarity(original, deciphered):
    result = ''
    count = 0
    for c in range(len(original)):
        if original[c] == deciphered[c]:
            result += original[c]
            count += 1
        else:
            result += '*'
            
    print()
    print('ПРОЦЕНТ ВЕРНОЙ РАСШИФРОВКИ :', count / len(list(original)))
    print()
    print('Верная часть расшифровки:')
    print()
    print(result)
```

```python
print(test_sent)
print()

chiper = np.random.permutation(list(char_occur.keys()))
chiper_dict = dict(zip(char_occur.keys(), chiper))

print('ШИФРОВКА')
chiper_text = ''.join([chiper_dict[c] for c in test_sent])
print(chiper_text)

deciphered = de_cipher_simple(chiper_text, char_to_count)

visual_similarity(test_sent, deciphered)
```

```python
def transition_dict(text, char_to_count):
    transition = dict(zip(char_to_count.keys(), [{}]*len(char_to_count)))
    for key_1 in transition.keys():
        for key_2 in transition.keys():
            transition[key_1][key_2] = 0
    for line in tqdm.tqdm(text):
        for char_i in range(len(line)-1):
                transition[line[char_i]][line[char_i+1]] += 1
    buff = {}
    for key in transition.keys():
        buff[key] = sum(transition[key].values())
        transition[key] = dict(zip(transition[key].keys(), np.array(list(transition[key].values())) / buff[key]))
        
    return transition
    
```

```python
trans_dic = transition_dict(text, char_to_count)
```

```python
def calc_llh(text, trans_dic):
    llh = 0
    for i in range(len(text) - 1):
        llh += np.log(trans_dic[text[i]][text[i + 1]])
        
    return llh
```

Как я понял, основная иддея метода в этом  случае - случайно блуждание по буквам (как вероятности переходы мы принмиаем частоты встреч буквы н+1 после буквы н), и если правдопобие чуть выросло - значит мы в вернмо направлении.

```python
def decihper(ciper_text, trans_dic, old_llh, iters = 10):
    for _ in range(iters):
        new_seq = np.random.permutation(list(trans_dic.keys()))
        new_seq = list(new_seq)[:7] + list(trans_dic.keys())[7:]
        new_dict = dict(zip(list(trans_dic.keys()), new_seq))
        
      
        new_text = []
        for i in range(len(ciper_text)):
            new_text.append(new_dict[ciper_text[i]])
        new_text = ''.join(new_text)
        llh = calc_llh(new_text , trans_dic)
        print(llh)
        if llh > old_llh or random.random() < np.exp(llh - old_llh):
            old_llh = llh
            ciper_text = new_text
            print(old_llh)
    return ciper_text
 
```
