```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from gensim.parsing.preprocessing import strip_punctuation
```

```python
text = [strip_punctuation(t).lower().replace('\n', '').replace('\t', '').replace('“', ' ').replace('„', ' ').replace('…', '').replace('»', '').replace('—', '').replace('«', '').replace('–', '') for t in text ]
```

```python
cnt = CountVectorizer(analyzer='char')
coc = cnt.fit_transform(text)
char_occur = dict(zip(cnt.get_feature_names(), coc.sum(axis=0).tolist()[0]))
```

```python
char_occur = dict(list(filter(lambda x: x[0] == ' ' or (ord(x[0]) > 900 and x[0] not in ['?', 'ї', '“', '„','№', '’', '—', '–', '.', ',']), sorted(list(char_occur.items()), key = lambda x: x[1]))))
```

```python
test_sent = 'Вечером князь Андрей и Пьер сели в коляску и поехали в Лысые Горы. Князь Андрей, поглядывая на Пьера, прерывал изредка молчание речами, доказывавшими, что он находился в хорошем расположении духа. Он говорил ему, указывая на поля, о своих хозяйственных усовершенствованиях. Пьер мрачно молчал, отвечая односложно, и казался погруженным в свои мысли. Пьер думал о том, что князь Андрей несчастлив, что он заблуждался, что он не знает истинного света и что Пьер должен прийти на помощь ему, просветить и поднять его. Но как только Пьер придумывал, как и что он станет говорить, он предчувствовал, что князь Андрей одним словом, одним аргументом уронит все его ученье, и он боялся начать, боялся выставить на возможность осмеяния свою любимую святыню.'
print(test_sent)
print()

test_sent_char = list(test_sent.lower())
i = 0
while i < len(test_sent_char):
    if test_sent_char[i] not in char_occur.keys():
        test_sent_char.pop(i)
    i += 1

test_sent = ''.join(test_sent_char)
print(test_sent)
```

```python
def de_cipher_simple(text, char_to_count):
    cnt = CountVectorizer(analyzer='char')
    coc = cnt.fit_transform([text])
    char_occur = dict(zip(cnt.get_feature_names(), coc.sum(axis=0).tolist()[0])) 
    sorted_sec = [x[0] for x in sorted(list(char_occur.items()), key = lambda x: x[1])][::-1]
    de_cipher = dict(zip(sorted_sec, list(char_to_count.keys())[::-1][:len(sorted_sec)]))
    res = ''.join([de_cipher[x] for x in text])
    print()
    print('РАСШИФРОВКА')
    print(res)
    return res
```

```python
def visual_similarity(original, deciphered):
    result = ''
    count = 0
    for c in range(len(original)):
        if original[c] == deciphered[c]:
            result += original[c]
            count += 1
        else:
            result += '*'
            
    print()
    print('ПРОЦЕНТ ВЕРНОЙ РАСШИФРОВКИ :', count / len(list(original)))
    print()
    print('Верная часть расшифровки:')
    print()
    print(result)
```

```python
print(test_sent)
print()

chiper = np.random.permutation(list(char_occur.keys()))
chiper_dict = dict(zip(char_occur.keys(), chiper))

print('ШИФРОВКА')
chiper_text = ''.join([chiper_dict[c] for c in test_sent])
print(chiper_text)

deciphered = de_cipher_simple(chiper_text, char_to_count)

visual_similarity(test_sent, deciphered)
```

```python
ok_chars = []
for item in char_to_count.items():
    key = item[0]
    if (ord(key[0]) > 900 or key[0] == ' ') and (ord(key[1]) > 900 or key[1] == ' '):
        ok_chars.append(item)
```

```python
char_occur = dict(ok_chars)
```

```python
def de_cipher_advanced(text, char_to_count):
    cnt = CountVectorizer(analyzer='char', ngram_range=(2, 2))
    coc = cnt.fit_transform([text])
    char_occur = dict(zip(cnt.get_feature_names(), coc.sum(axis=0).tolist()[0])) 
    sorted_sec = [x[0] for x in sorted(list(char_occur.items()), key = lambda x: x[1])][::-1]
    de_cipher = dict(zip(sorted_sec, list(char_to_count.keys())[::-1][:len(sorted_sec)]))
    res = ''.join([de_cipher[text[i:i+2]] for i in range(0, len(text), 2)])
    print()
    print('РАСШИФРОВКА')
    print(res)
    return res
```

```python
print(test_sent)
print()

chiper = np.random.permutation(list(char_occur.keys()))
chiper_dict = dict(zip(char_occur.keys(), chiper))

print('ШИФРОВКА')
chiper_text = ''.join([chiper_dict[test_sent[i:i+2]] for i in range(0, len(test_sent), 2)])
print(chiper_text)

deciphered = de_cipher_advanced(chiper_text, char_to_count)

visual_similarity(test_sent, deciphered)
```
